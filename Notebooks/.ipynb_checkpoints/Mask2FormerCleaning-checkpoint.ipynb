```python
import os
import numpy as np
import torch
import json
import detectron2
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw, ImageFont, UnidentifiedImageError
from detectron2.data import MetadataCatalog
from detectron2.utils.visualizer import Visualizer, ColorMode
from transformers import MaskFormerImageProcessor, Mask2FormerForUniversalSegmentation, Mask2FormerConfig, Mask2FormerModel
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
def panoptic(seg, seg_info, image):

    metadata = MetadataCatalog.get("coco_2017_val_panoptic")
    #print(metadata)
    thing_classes = metadata.thing_classes
    stuff_classes = metadata.stuff_classes

    for res in seg_info:
        res['category_id'] = res.pop('label_id')
        pred_class = res['category_id']
        score = res['score']
        isthing = pred_class in metadata.thing_dataset_id_to_contiguous_id.values()
        res['isthing'] = bool(isthing)

        if isthing:
            if pred_class < len(thing_classes):
                label_name = thing_classes[pred_class]
            else:
                label_name = 'Unknown'
        else:
            if pred_class < len(stuff_classes):
                label_name = stuff_classes[pred_class]
            else:
                label_name = 'Unknown'

        res['isthing'] = f"{label_name} ({score:.2f})"

    visualizer = Visualizer(np.array(image)[:, :, ::-1], metadata=metadata, instance_mode=ColorMode.IMAGE)

    out = visualizer.draw_panoptic_seg_predictions(seg.cpu(), seg_info, alpha=0.5)
    #print(seg_info)
    output_img = Image.fromarray(out.get_image())

    return output_img, seg_info

def predicted(img_path):

    pretrained = "facebook/mask2former-swin-large-coco-panoptic"

    model = Mask2FormerForUniversalSegmentation.from_pretrained(pretrained).to(torch.device(device))
    model.eval()

    image_processor = MaskFormerImageProcessor.from_pretrained(pretrained)

    image = Image.open(img_path)
    inputs = image_processor(images=image, return_tensors="pt").to(torch.device(device))

    with torch.no_grad():
        outputs = model(**inputs)

    #print(outputs)
    result = image_processor.post_process_panoptic_segmentation(outputs, target_sizes=[image.size[::-1]])[0]
    #print(result)
    output_result, seg_data = panoptic(result["segmentation"], result['segments_info'], image)

    return output_result, seg_data
```
